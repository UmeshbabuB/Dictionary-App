```python
'''This project uses the skills learnt from week 1 to week 6. The aim of this project is to create an English Dictionary app 
that returns definitions of English words. When the application is started, it displays a menu as follows:

Main Menu
    1.Add a new word
    2.Find the meaning
    3.Update a word
    4.Exit
        
        Enter Choice:

If item 1 is selected, it should prompt for a word and then accept its meaning. This pair of word and its meaning 
should be stored in a file called ‘words.txt’. 

When item 2 is selected, it should prompt for a word, search that word 
in the file ‘words.txt’ and return the meaning if it is found. If the meaning is not found, it should display an 
appropriate message. 

Similarly, when choice 3 is entered, it should accept a word and update its meaning.

In all the above three cases, the prompt should return to the main menu. The words and their meanings should be stored
as a dictionary in words.txt, using the concept of serialization. 

If choice 4 is selected, a graceful exit should be performed.'''
```

```python

d={}
while 1:
    
    import json
    print('*****ENGLISH DICTIONARY APP*****')
    print('''Main Menu
      1.Add a new word
      2.Find the meaning
      3.Update a word
      4.Exit
        
  Enter Choice:''')
    
    n=int(input('Choice  :'))
    
    print()
    
    if n==1:
        word=input('Enter the word  :')
        meaning=input('Enter the meaning for word  :')
        file_1=open('words.txt','w')
        d[word]=meaning
        json.dump(d,file_1)
        file_1.close()
        print('***-----------------------------***')
        print()
        continue
        
    if n==2:
        j=input('Enter the Word  :')
        for i in d:
            if i==j:
                print('Meaning is founded for the word that your searching in the dictionary','"',j,'"',':',d[i])
                break
        else:
            print('Meaning for the word you are searching','"',j,'"','is not founded in the dictionary')
        
        print('***-------------------------------***')
        print()
        continue
        
    if n==3:
        word=input('Enter the word :')
        meaning=input('Enter the meaning to update  :')       
        file_1=open('words.txt','r')
        f1=json.load(file_1)
        file_1.close()
        print(d)
        for i in d:
            if i==word:
                d[i]=meaning
                print(d)
                break
                
        else:
            file_1=open('words.txt','w')
            d[word]=meaning
            f1=json.dump(d,file_1)     
            file_1.close()
            print('After udating the meaning for the word you entered',d) 
        
        print('***-------------------------------***')
        print()
        continue
        
    if n==4:
        print('Exit')
        break   
        
```
